```python
import pandas as pd
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
from collections import Counter

# Load training data
df = pd.read_csv("Fact_BH_Val_Use_New.csv")

# Encode class labels
label_encoder = LabelEncoder()
df['class_encoded'] = label_encoder.fit_transform(df['Remark'])

# Drop non-numeric columns
X = df.drop(columns=['Remark', 'class_encoded'], errors='ignore')
X = X.select_dtypes(include='number')
y = df['class_encoded']

# Check class distribution before balancing
print("Original class distribution:", Counter(y))

# Split before applying SMOTE to avoid data leakage
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42)

# Apply SMOTE to training set only
smote = SMOTE(random_state=42)
X_train_bal, y_train_bal = smote.fit_resample(X_train, y_train)

# Check class distribution after SMOTE
print("Balanced class distribution:", Counter(y_train_bal))

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_bal)
X_test_scaled = scaler.transform(X_test)

# Train SVM model
svm_model = SVC(kernel='rbf', class_weight='balanced', probability=True, random_state=42)
svm_model.fit(X_train_scaled, y_train_bal)

# Evaluate on test set
y_pred = svm_model.predict(X_test_scaled)
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nPredicted classes in test set:", set(y_pred))

# ----------- PREDICTION ON FISHNET DATA ----------- #
# Load fishnet data
fishnet_df = pd.read_csv("Factr_Fish_Val_Use_New_Crct.csv")
X_fishnet = fishnet_df.select_dtypes(include='number')

# Scale using same scaler
X_fishnet_scaled = scaler.transform(X_fishnet)

# Predict class and probabilities
fishnet_df['svm_pred_class'] = svm_model.predict(X_fishnet_scaled)
probs = svm_model.predict_proba(X_fishnet_scaled)
for i, class_name in enumerate(label_encoder.classes_):
    fishnet_df[f'svm_prob_{class_name.lower()}'] = probs[:, i]

# Save results
fishnet_df.to_csv("fishnet_svm_predictions_fixed.csv", index=False)
print("✔️ Fishnet predictions with SMOTE-balanced SVM saved.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc, roc_auc_score
from sklearn.model_selection import train_test_split

# Load your CSV
df = pd.read_csv("Data_4_Val_SVM.csv")

# Extract actual and predicted probabilities
y_true = df["Actual"]
classes = sorted(y_true.unique())  # Example: [0, 1, 2]
n_classes = len(classes)

# Binarize the actual classes
y_test_bin = label_binarize(y_true, classes=classes)

# Predicted probabilities: Ensure your CSV has Prob_0, Prob_1, ..., Prob_n
y_score = df[[f'Prob_{cls}' for cls in classes]]

# Compute ROC curve and AUC for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_score.iloc[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Compute macro-average ROC curve and AUC
fpr["macro"], tpr["macro"], _ = roc_curve(y_test_bin.ravel(), y_score.values.ravel())
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

# Plot
plt.figure(figsize=(10, 6))
colors = ['blue', 'green', 'red', 'orange', 'purple']

for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2,
             label=f"Class {classes[i]} (AUC = {roc_auc[i]:.2f})")

# Macro-average ROC
plt.plot(fpr["macro"], tpr["macro"], color='navy', linestyle='--',
         label=f"Macro-average (AUC = {roc_auc['macro']:.2f})", linewidth=2)

plt.plot([0, 1], [0, 1], 'k--', lw=1)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Multiclass ROC Curve')
plt.legend(loc="lower right")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# fine tuning xgboost model
import pandas as pd
from xgboost import XGBClassifier
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

# Load and encode data
df = pd.read_csv("Fact_BH_Val_Use_New.csv")
label_encoder = LabelEncoder()
df['class_encoded'] = label_encoder.fit_transform(df['Remark'])

X = df.drop(columns=['remarks', 'class_encoded'], errors='ignore').select_dtypes(include='number')
y = df['class_encoded']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split for grid search validation
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42, stratify=y)

# Define the parameter grid for GridSearchCV
param_grid = {
    'n_estimators': [100, 200],  # Number of boosting rounds
    'max_depth': [3, 5, 7],  # Maximum depth of a tree
    'learning_rate': [0.05, 0.1, 0.2],  # Learning rate
    'subsample': [0.7, 1.0],  # Fraction of samples used for each tree
    'colsample_bytree': [0.7, 1.0]  # Fraction of features used for each tree
}

# Initialize the XGBoost model
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss', random_state=42)

# Initialize GridSearchCV
grid_search = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    scoring='roc_auc_ovr',  # ROC AUC for multiclass
    cv=3,  # 3-fold cross-validation
    verbose=1,
    n_jobs=-1
)

# Fit the model to the training data
grid_search.fit(X_train, y_train)

# Best parameters found by GridSearchCV
print("✅ Best parameters found:", grid_search.best_params_)

# Use the best model to predict on the test set
best_model = grid_search.best_estimator_
y_probs = best_model.predict_proba(X_test)

# Calculate ROC AUC score for multiclass
auc_score = roc_auc_score(y_test, y_probs, multi_class='ovr', average='micro')
print("📈 Micro-averaged ROC AUC after tuning:", round(auc_score * 100, 2), "%")

# Now predicting on the fishnet data
fishnet_df = pd.read_csv("Factr_Fish_Val_Use_New_Crct.csv")
X_fishnet = fishnet_df.select_dtypes(include='number')
X_fishnet_scaled = scaler.transform(X_fishnet)

# Predict class and probabilities for fishnet data
fishnet_df['xgb_pred_class'] = best_model.predict(X_fishnet_scaled)
probs = best_model.predict_proba(X_fishnet_scaled)

# Assign probabilities to respective class columns
for i, class_name in enumerate(label_encoder.classes_):
    fishnet_df[f'xgb_prob_{class_name.lower()}'] = probs[:, i]

# Save the predictions to a new CSV file
fishnet_df.to_csv("fishnet_xgb_predictions_tuned.csv", index=False)
```

```python
# import the necessary libraries

import numpy as np
import pandas as pd

# Read the pairwise comparison matrix as a CSV file and convert to numpy array

pairwise_matrix = pd.read_csv('Consolidated_AHP_Pairwise_Matrix.csv', index_col=0).values

# Square the Pairwise Comparison Matrix
squared_matrix = np.dot(pairwise_matrix, pairwise_matrix)

# Normalize the Squared Pairwise Comparison Matrix
column_sums = np.sum(squared_matrix, axis=0)  # Column-wise sum
normalized_matrix = squared_matrix / column_sums  # Normalize each column

# Compute Row Sums of the Normalized Matrix
row_sums = np.sum(normalized_matrix, axis=1)

# Calculate Criteria Weights
criteria_weights = row_sums / np.sum(row_sums)

# Consistency Check
# Multiply the original matrix by the weights vector
weighted_sum_vector = np.dot(pairwise_matrix, criteria_weights)

# Compute the consistency vector
consistency_vector = weighted_sum_vector / criteria_weights

# Calculate lambda_max
lambda_max = np.mean(consistency_vector)

# Number of criteria (n) and Random Index (RI) values
n = pairwise_matrix.shape[0]
RI_values = {1: 0, 2: 0, 3: 0.58, 4: 0.9, 5: 1.12, 6: 1.24, 7: 1.32, 8: 1.41, 9: 1.45, 10: 1.49}
RI = RI_values.get(n, 1.49)  # Default to max RI if n > 10

# Consistency Index (CI) and Consistency Ratio (CR)
CI = (lambda_max - n) / (n - 1)
CR = CI / RI if RI != 0 else 0

# Output results
print("Pairwise Comparison Matrix (Original):")
print(pairwise_matrix)
print("\nSquared Pairwise Comparison Matrix:")
print(squared_matrix)
print("\nNormalized Matrix:")
print(normalized_matrix)
print("\nCriteria Weights:")
print(criteria_weights)
print(f"\nConsistency Ratio (CR): {CR:.4f}")
```

```python
# converting the fishnet values into rating
# rating needed for further computations of AHP

import pandas as pd

# Load the fishnet input
df = pd.read_csv("Factr_Fish_Val_Use_New_Crct.csv")

# Rating functions using actual value intervals
def rate_twi(x):
    if -0.782 <= x <= 3.371: return 1
    elif 3.371 <= x <= 5.41: return 2
    elif 5.41 <= x <= 8.733: return 3
    elif 8.733 <= x <= 18.475: return 4
    else: return 1

def rate_rainfall(x):
    if 2104 <= x <= 2339: return 1
    elif 2339 <= x <= 2446: return 2
    elif 2446 <= x <= 2579: return 3
    elif 2579 <= x <= 2767: return 4
    else: return 1

def rate_prox_surface(x):  # in meters (distance to surface water)
    if 0 <= x <= 0.022: return 1
    elif 0.023 <= x <= 0.047: return 2
    elif 0.047 <= x <= 0.079: return 3
    elif 0.079 <= x <= 0.145: return 4
    else: return 1

def rate_lineament_density(x):
    if 0 <= x <= 0.090: return 1
    elif 0.090 <= x <= 0.270: return 2
    elif 0.270 <= x <= 0.500: return 3
    elif 0.500 <= x <= 0.860: return 4
    else: return 1


def rate_geology(x):
    # x must already be encoded as numeric values representing geologic types
    geology_ratings = {
        1: 2,  
        2: 3,  
        3: 4,  
        4: 3,  
        5: 1,  
        6: 2,  
        7: 3   
    }
    return geology_ratings.get(int(x), 1)

def rate_drainage_density(x):
    if 0.030 <= x <= 0.750: return 4
    elif 0.750 <= x <= 1.060: return 3
    elif 1.060 <= x <= 1.380: return 2
    elif 1.380 <= x <= 2.330: return 1
    else: return 1

def rate_ndvi(x):
    if 0.0184 <= x <= 0.2684: return 1
    elif 0.2684 <= x <= 0.4723: return 2
    elif 0.4723 <= x <= 0.6993: return 3
    elif 0.6993 <= x <= 0.993: return 4
    else: return 1

def rate_aspect(x):
    # assuming aspect is in degrees: 0–360
    if x == -1: return 4   # North
    elif -1 <= x <= 22.5: return 4             # NE
    elif 22.5 <= x <= 67.5: return 3            # East
    elif 67.5 <= x <= 112.5: return 2            # SE
    elif 112.5 <= x <= 157.5: return 1            # South
    elif 157.5 <= x <= 202.5: return 1            # SW
    elif 202.5 <= x <= 247.5: return 1
    elif 247.5 <= x <= 292.5: return 2             # West
    elif 292.5 <= x <= 337.5: return 3
    elif 337.5 <= x <= 360: return 4
    else: return 1

def rate_slope(x):
    if 0 <= x <= 4.0: return 4
    elif 4.0 <= x <= 10.0: return 3
    elif 10.0 <= x <= 19.0: return 2
    elif 19.0 <= x <= 65: return 1
    else: return 1

# Apply ratings
df['TWI_Rating'] = df['TWI'].apply(rate_twi)
df['Rainfall_Rating'] = df['RAINFALL'].apply(rate_rainfall)
df['ProxSW_Rating'] = df['PTWBDS'].apply(rate_prox_surface)
df['LineamentDensity_Rating'] = df['LD'].apply(rate_lineament_density)
df['Geology_Rating'] = df['GEOLOGY'].apply(rate_geology)
df['DrainageDensity_Rating'] = df['DD'].apply(rate_drainage_density)
df['NDVI_Rating'] = df['NDVI'].apply(rate_ndvi)
df['Aspect_Rating'] = df['ASPT'].apply(rate_aspect)
df['Slope_Rating'] = df['SLOPE'].apply(rate_slope)

# Save output with ratings
df.to_csv("fishnet_rated_output.csv", index=False)
print("✅ Rating complete and saved to 'fishnet_rated_output.csv'")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize
from sklearn.metrics import roc_curve, auc

# Load combined predictions CSV
df = pd.read_csv("combined_predictions.csv")

# Define your custom natural breaks for AHP model class assignment
def classify_ahp(index):
    if index <= 2.004:
        return 0  # Low
    elif index <= 2.316:
        return 1  # Moderate
    else:
        return 2  # High

# Apply classification to AHP index column
df['ahp_pred'] = df['ahp_index'].apply(classify_ahp)

# Binarize the actual class
classes = [0, 1, 2]
y_true = label_binarize(df['actual_class'], classes=classes)

# Dictionary of model predictions
model_preds = {
    "SVM": df["svm_pred"],
    "RF": df["rf_pred"],
    "XGBoost": df["xgb_pred"],
    "AdaBoost": df["ada_pred"],
    "AHP": df["ahp_pred"]
}

# Plotting micro-averaged ROC
plt.figure(figsize=(8, 6))

for model_name, y_pred in model_preds.items():
    y_pred_bin = label_binarize(y_pred, classes=classes)
    fpr, tpr, _ = roc_curve(y_true.ravel(), y_pred_bin.ravel())
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, lw=2, label=f'{model_name} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=1)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Micro-Averaged ROC Curve (ML and AHP Models)")
plt.legend(loc="lower right")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import label_binarize
from sklearn.metrics import (
    roc_curve, auc, accuracy_score, f1_score,
    classification_report, precision_score, recall_score
)

# Load CSV
df = pd.read_csv("all_ml_models_val_data.csv")

# True labels and number of classes
y_true = df["True_Class"].values
classes = sorted(df["True_Class"].unique())
n_classes = len(classes)

# Binarize true labels for micro-average ROC
y_true_bin = label_binarize(y_true, classes=classes)

# Define models and their probability column patterns
models = {
    "SVM": [f"svm_prob_{i}" for i in classes],
    "RF": [f"rf_prob_{i}" for i in classes],
    "XGBoost": [f"xgb_prob_{i}" for i in classes],
    "AdaBoost": [f"ada_prob_{i}" for i in classes],
    "AHP": None  # special handling for AHP
}

# Store evaluation results
roc_curves = {}
metrics_summary = {}

# Loop through models
for model_name, prob_cols in models.items():
    if model_name != "AHP":
        y_score = df[prob_cols].values
    else:
        # For AHP, repeat the index as a 2D array with shape (n_samples, n_classes)
        y_score = np.tile(df["ahp_index"].values.reshape(-1, 1), (1, n_classes))

    # Compute micro-average ROC
    fpr, tpr, _ = roc_curve(y_true_bin.ravel(), y_score.ravel())
    roc_auc = auc(fpr, tpr)
    roc_curves[model_name] = (fpr, tpr, roc_auc)

    # Compute predicted class
    y_pred = np.argmax(y_score, axis=1)

    # Classification metrics
    metrics_summary[model_name] = {
        "Accuracy": accuracy_score(y_true, y_pred),
        "F1 (macro)": f1_score(y_true, y_pred, average='macro'),
        "Precision (macro)": precision_score(y_true, y_pred, average='macro'),
        "Recall (macro)": recall_score(y_true, y_pred, average='macro'),
    }

# ---------------------------
# 📈 Plot all ROC curves
# ---------------------------
plt.figure(figsize=(10, 8))
for model_name, (fpr, tpr, roc_auc) in roc_curves.items():
    plt.plot(fpr, tpr, lw=2, label=f"{model_name} (AUC = {roc_auc:.2f})")

plt.plot([0, 1], [0, 1], color='grey', lw=1, linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Micro-Averaged ROC Curve: ML Models vs AHP")
plt.legend(loc="lower right")
plt.grid(True)
plt.tight_layout()
plt.show()

# ---------------------------
# 📋 Print classification metrics
# ---------------------------
print("\nClassification Performance Metrics (per model):\n")
for model_name, scores in metrics_summary.items():
    print(f"🔹 {model_name}")
    for metric_name, value in scores.items():
        print(f"   {metric_name}: {value:.2f}")
    print()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from scipy.stats import pearsonr
import numpy as np

# === Step 1: Load your CSV file ===

df = pd.read_csv("ahp_vs_bh_indices.csv")

# Check column names
if not {'BH_Index_Value', 'AHP_Index_Value'}.issubset(df.columns):
    raise ValueError("CSV must contain columns 'BH_Index_Value' and 'AHP_Index_Value'")

# === Step 2: Compute metrics ===
actual = df['BH_Index_Value']
predicted = df['AHP_Index_Value']

r2 = r2_score(actual, predicted)
rmse = mean_squared_error(actual, predicted, squared=False)
mae = mean_absolute_error(actual, predicted)
corr, _ = pearsonr(actual, predicted)

# === Step 3: Plot with metrics ===
plt.figure(figsize=(8, 6))
plt.scatter(actual, predicted, color='dodgerblue', alpha=0.6, label="Data Points")
m, b = np.polyfit(actual, predicted, 1)
plt.plot(actual, m*actual + b, color='red', linewidth=2, label="Regression Line")
plt.xlabel("Actual Borehole Index Value")
plt.ylabel("Predicted AHP Index Value")
plt.title("AHP vs Borehole Index\n"
          f"R² = {r2:.3f} | RMSE = {rmse:.3f} | MAE = {mae:.3f} | r = {corr:.3f}")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
